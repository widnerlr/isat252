```python
import pandas
```

```python
df = pandas.read_excel('s3://isat252-widner/house_price.xls')
df[:10]
df.plot(x is 'built_in',y is 'prices')
```

# Line Chart 
avg prices per year 

```python
avg_price_per_year = df.groupby('built_in').mean()['price']
avg_price_per_year.plot() #plot creates line chart by default 
```

## Histogram 
to show the distribution of the prices 

```python
df['price'].hist() #all values of a single column
```

most houses are around 200k 

## bar chart 
avg prices per house type

```python
avg_prices_per_house_type = df.groupby('house_type').mean()['price'] 
#avg_prices_per_house_type[:]
avg_prices_per_house_type.plot.bar() #return a bar chart 
```

single family houses have the highest price 

## pie chart 
show number of records for house type 


```python
count_per_house_type = df['house_type'].value_counts()
#count_per_house_type[:]
count_per_house_type.plot.pie() #create a pie chart 
```

## Scatterplot 
show relationship between price and area 

```python
df.plot.scatter(x ='area',y='price')
```
